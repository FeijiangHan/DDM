```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
d=dict()
d['subset distribution']=['standard','skew']*3
d['Accuracy']=[0.764,0.991,0.525,0.951,0.817,0.915]
d['dataset']=['IMDB-B']*2+['IMDB-M']*2+['COLLAB']*2
data=pd.DataFrame(d)
```

```python
sns.set_palette("pastel")
sns.set_context("paper", font_scale=2)
plt.figure(figsize=(6,4),dpi=300)
ax=sns.barplot(x='dataset',y='Accuracy',hue='subset distribution',data=data,width=0.5)
plt.legend(loc='upper right', bbox_to_anchor=(1, 1.2),fontsize=10)
sns.despine(ax=ax, top=True, right=True, left=False, bottom=False)

for p in ax.patches:
    ax.annotate(format(p.get_height(), '.3f'), 
                   (p.get_x() + p.get_width() / 2., p.get_height()), 
                   ha = 'center', va = 'center', 
                   xytext = (0, 10), 
                   textcoords = 'offset points',
               fontsize=8)
plt.savefig('skew.pdf',dpi=300,bbox_inches="tight")
```
