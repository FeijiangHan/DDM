```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = "record.json"
with open(file, encoding="utf-8") as f:  # 注意编码要和文件编码一致，不加encoding参数默认使用gbk编码读取文件
    ddm = json.load(f)
```

```python
file = "record_dm.json"
with open(file, encoding="utf-8") as f:  # 注意编码要和文件编码一致，不加encoding参数默认使用gbk编码读取文件
    dm = json.load(f)
```

```python
import pandas as pd
```

```python
dm = pd.DataFrame(dm)
dm['category'] = 'white noise'
ddm = pd.DataFrame(ddm)
ddm['category'] = 'directional noise'
```

```python
data = pd.concat([ddm, dm])
```

```python
data
```

```python
sns.lineplot(data=data, x='epoch', y='loss', hue='category')
```

```python
plt.figure(figsize=(6,4),dpi=300)
sns.lineplot(data=data, x='epoch', y='loss', hue='category')
# sns.despine()
# plt.axis([0,1000,0,2.1])
plt.legend(loc='upper right')
plt.ylabel('Training Loss')
plt.savefig('AmazonCoBuyPhotoDataset_Training_Loss_curve.pdf',dpi=300,bbox_inches="tight")
```
